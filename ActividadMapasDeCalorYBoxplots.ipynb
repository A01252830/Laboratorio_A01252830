```python
credits = pd.read_csv("credit.csv")
```

----------------------------------------------------------------------------------------------------------------------------------------------------------------
#**Boxplots:**

```python
fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Income")
plt.show()

fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Limit")
plt.show()

fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Rating")
plt.show()

fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Cards")
plt.show()
```

```python
fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Age")
plt.show()

fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Education")
plt.show()

fig = plt.figure(figsize=(20,3))
sns.boxplot(data=credits, x="Balance")
plt.show()
```

----------------------------------------------------------------------------------------------------------------------------------------------------------------
#**Histogramas:**

```python
fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Income")
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Limit")
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Rating")
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Cards")
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Age")
plt.grid(True)
plt.show()

fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Education")
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
sns.histplot(data=credits, x="Balance")
plt.grid(True)
plt.show()
```

#**Mapa de Calor:**

```python
plt.figure(figsize=(8, 7))
creditcorr = credits.corr()
sns.heatmap(data=creditcorr, annot=True)
```

¿Hay alguna variable que no aporta información?
Las variables que no importan información son Gender, Student, Married y Ethnicity, esto se debe a que no son variables numéricas sino variables cualitativas.

Si tuvieras que eliminar variables, ¿cuáles quitarías y por qué?
Quitaría Student o Married porque los datos que tienen no son muy importantes. Pero si fueran datos numéricos, quitaría Cards porque no da muchos datos para los gráficos.

¿Existen variables que tengan datos extraños?
Los datos de Boxplots para Age, Education y Balance no tienen datos atípicos, lo cual es extraño porque las otras variables tienen datos atípicos, o al por lo menos en el lado máximo.

Si comparas las variables, ¿todas están en rangos similares? ¿Crees que esto afecte?
Casi todas las variables en todos los datos tienen diferentes rangos, esto significa que cada variable tiene datos únicos que son útiles para obtener información. Las únicas variables que son casi iguales son Cards y Education.

¿Puedes encontrar grupos qué se parezcan? ¿Qué grupos son estos?
En Boxplots, las variables Age, Education y Balance se parecen porque no tienen valores atípicos, y las variables Limit y Rating se parecen por su posición, longitud y valores atípicos. Otro que yo consideraría similar es en los histogramas con Limit y Rating debido a cómo aumenta y disminuye.
